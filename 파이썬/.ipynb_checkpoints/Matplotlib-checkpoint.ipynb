```python
# 라이브러리 import
import matplotlib.pyplot as plt
```

```python
# plot 함수 - 일직선으로 된 line plot 그리기
y = [2,4,6]
plt.plot(y) # 그래프 그리기
plt.show()  # 그래프 출력
```

```python
x = [1,2,3]
y = [2,4,6] 

plt.plot(x,y)
plt.show()
```

```python
# linestyle
# 그려지는 line의 스타일을 지정하는 속성 -> ls
plt.plot(x,y,ls='--')
plt.show()
```

```python
x=[2,4,6,8]
y=[10,13,15,20]
y2=[10,20,17,15]

plt.plot(x,y)
plt.plot(x,y2, ls=':')
plt.show()
```

```python
# marker
plt.plot(x,y,marker='*')
plt.show()
```

```python
# line width
# 라인의 두께 지정하기
plt.plot(x,y,linewidth = 30)
plt.show()
```

```python
# color 색상 변경
plt.plot(x,y, color='#008080',marker='o')
plt.show()
```

```python
# marker face color
# 마커의 색상 지정 -> mfc
plt.plot(x,y, color='#008080',marker='o',mfc='yellow')
plt.show()
```

### 차대차 사건중 죽거나 다친 사람이 많은 발생지시도 알아보기

```python
# 27개 컬럼 보이게 하기
pd.set_option( 'display.max_columns',27 )
```

```python
# 1. 차대차 사건 중
# 2. (1)죽거나 다친 사람이 많은 (2)발생지시도
# 3. 시각화
data
```

```python
# 1. 차대차 사건 중
# 불리언 인덱싱
# 사고유형_대분류 컬럼으로 '차대차' 인 데이터 가져오기
data['사고유형_대분류']=='차대차' # 불리언
acc = data[data['사고유형_대분류']=='차대차']
```

```python
acc
```

```python
# 2. (1)죽거나 다친 사람(사상자수)이 많은 (2)발생지시도
# 사상자수 컬럼, 
acc[['사상자수','발생지시도']] # = acc,loc[:,['사상자수','발생지시도']]
```

```python
acc['발생지시도'].value_counts()
# 시도별 사상자수 파악 힘듬
```

```python
# groupby()
# 같은 값끼리 묶어 통계 연산 값을 보여줌
acc_result=acc.loc[:,['사상자수','발생지시도']].groupby('발생지시도').sum()
```

```python
x = acc_result.index
```

```python
y = acc_result['사상자수'].values
```

```python
# 3. 그래프 그리기
plt.figure(figsize=(10,5)) # (가로,세로)
plt.bar(x,y,color='b')
plt.title('차vs차 교통사고의 사상자수')
plt.xlabel('지역')
plt.ylabel('사상자 수')
plt.show()
```
