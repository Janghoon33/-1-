### 인덱싱과 슬라이싱

```python
# 문자열 인덱싱

s4 = "My name is JY"
print(s4[0])
print(s4[8])
print(s4[12])
print(s4[-1])
```

```python
print(s4[3]+s4[4]+s4[5]+s4[6])
```

```python
# 문자열 슬라이싱
# ->0을 포함하고 4보다 작을때 까지 출력


s5 = "My name is JH"

# 0부터 잘라내고 싶을 때 
print(s5[:4])
print(s5[3:])
print(s5[-5:])
```

```python
day="2020년 3월 3일의 날씨는 맑음입니다."

print("날짜 : "+day[0:11])
print("날씨 : "+day[-6:-4])


print(s5[0:5])
print(s5[3:10])
print(s5[3:13])
print(s5[-5:-1])
```

```python
month = 3
day = 23
s6 = "오늘은"+str(month)+"월"+str(day)+"일 입니다."
s6
```

```python
# %기호 포매팅 -> %d를 활용해서 정수 대입

month = 3
day = 23
s6 = "오늘은 %d월 %d일 80%%입니다."%(month,day)
print(s6)
```

```python
# format함수 포매팅
month = 3
day = 23
s6 = "오늘은 {}월 {}일 입니다.".format(month,day)
print(s6)
```

```python
# f문자열 포매팅
month = 3
day = 23
s6 = f"오늘은 {month}월 {day}일 입니다"
print(s6)
```

```python
x = 100
y = 200
sum2 = x + y
s7 = f"{x}와 {y}의 합은 {sum2}입니다."
print(s7)
```

### 문자열 관련 함수

```python
s7 = "My : name : is : Janghoon"
# 문자열에 포함된 문자 개수 세기
print(s7.count("o"))
print(s7.count("Janghoon"))
```

```python
s7 = "My : name : is : Janghoon"
# 문자 위치 알려주기
print(s7.find("n"))
```

```python
s7 = "My : name : is : Janghoon"
# 대소문자 변환
print(s7.upper())
print(s7.lower())
```

```python

# 문자열 바꾸기
print(s7.replace("Janghoon","JeonJangHoon"))
print(s7.rfind("e"))
```

```python
s7 = "My : name : is : Janghoon"
# 문자열 자르기
print(s7.split(":"))
```
