# 네이버 페이지에서 이미지 데이터 수집

```python
from selenium import webdriver as wb
from bs4 import BeautifulSoup as bs
import time
# 파일시스템을 위한 라이브러리 ex) 파일, 폴더를 생성,삭제
import os
# 이미지의 경로를 파일로 변환시켜주는 라이브러리
from urllib.request import urlretrieve
```

```python
# 폴더 생성
# 터미널 명령
# mk : 만들기
# rm : 지우기
# cd : 경로변경
# ls : 파일확인
# ed : 편집
if not os.path.isdir("C:/Users/smhrd/Desktop/이미지") : 
    os.mkdir("C:/Users/smhrd/Desktop/이미지")
```

```python
driver = wb.Chrome()
driver.get("https://search.naver.com/search.naver?where=image&sm=tab_jum&query=%EA%BC%AC%EB%B6%80%EA%B8%B0")
```

```python
body = driver.find_element_by_css_selector("body")
for i in range(20) : 
    body.send_keys(Keys.END)
    time.sleep(1)
```

```python
soup = bs(driver.page_source,"lxml")
```

```python
img = soup.select("img._image._listImage")
```

```python
# 수집한 요소에서 컨텐츠가 아닌 속성을 추출하는 방법
img[0]["src"]
```

```python
imgList = []
for i in img : 
    imgList.append(i["src"])
imgList
```

```python
for i in range(40) : 
    urlretrieve(imgList[i],"C:/Users/smhrd/Desktop/이미지/"+str(i)+".jpg")
    time.sleep(0.5)
```
