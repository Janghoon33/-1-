```python
# requests
# - 서버에서 데이터를 요청해서 응답받기 위한 라이브러리
# - 요청하고, 응답 받으면 역할이 끝.
# - 브라우저의 역할을 대신
# - 페이지정보를 알아오는 함수 .get("url")
```

## Melon사이트 페이지 정보 요청

```python
# <Response [406]>
# 멜론서버에서 브라우저가 아닌 코드라고 인식했기때문에 접근이 불가능
req.get("http://www.melon.com")
```

```python
# 브라우저인척 속이는 작업
h = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/100.0.4896.127 Safari/537.36"}
```

```python
res = req.get("http://www.melon.com", headers = h)
```

```python
res.text
```

# 네이버페이지에서 블로그,뉴스 글자 수집!

```python
res = req.get("http://www.naver.com")
res.text
```

```python
# 3. 컴퓨터가 이해할 수 있는 html언어로 변형
# bs객체화를 진행
# beautifulSoup 라이브러리
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(res.text,"lxml")
```

```python
soup.select("a.nav")
```

```python
# 2. 뉴스,블로그 등과 같은 글자만 수집
# 2.1 html태그 정보를 알려주기!
# 필요한 요소를 검색해서 수집하는 명령 .select("선택자") ★
soup.select("a.nav")
# 반환받은 리스트 데이터 중에 첫번째 데이터만 가지고올때 select_one("선택자") ★
soup.select_one("a.nav")
```

```python
txt = soup.select("a.nav")
```

```python
for i in txt : 
    print(i.text)
```
